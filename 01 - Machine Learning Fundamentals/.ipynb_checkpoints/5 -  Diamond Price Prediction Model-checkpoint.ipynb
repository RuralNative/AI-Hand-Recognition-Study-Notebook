## Feature Engineering

```python
# Combine numerical features with high correlation with each other to reduce dimensionality
dataframe['dimension'] = dataframe['x']/dataframe['y']/dataframe['z']
```

```python
# Drop irrelevant numerical features
dataframe = dataframe.drop('Unnamed: 0', axis=1)
dataframe = dataframe.drop('x', axis=1)
dataframe = dataframe.drop('y', axis=1)
dataframe = dataframe.drop('z', axis=1)

dataframe.info()
```

```python
# Check potential features that may be produced upon object conversion
for column in object_columns:
    print(pd.get_dummies(dataframe[column]))
```

```python
# Use one-hot encoding for object features conversion
dataframe_encoded = pd.get_dummies(data=dataframe, columns=['cut','color','clarity'], dtype='float64')
dataframe_encoded.head()
```

```python
# Revisualize to check features correlation
dataframe_numerics = dataframe_encoded.select_dtypes(include = (np.number))
corr = dataframe_numerics.corr()
plt.figure(figsize=(24,24))
sns.heatmap(corr, annot=True)
plt.show()
```

```python
dataframe_encoded.info()
```

```python
# Check for NaN values
print(np.isnan(dataframe_encoded).sum())

# Check for infinite values
print(np.isinf(dataframe_encoded).sum())
```

```python
# Delete rows that have infinite values
dataframe_encoded = dataframe_encoded.replace([np.inf, -np.inf], np.nan)
dataframe_encoded = dataframe_encoded.dropna()
dataframe_encoded.info()
```

```python
# Check for null values
dataframe_encoded.isnull().sum()
```

```python
# Standardize/regularize data in the DataFrame
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(dataframe_encoded)
standardized_df = pd.DataFrame(scaled_data, columns=dataframe_encoded.columns)
```

```python
# Show headings of the preprocessed dataset
standardized_df.head()
```

## Training

```python
# Import necessary modules/packages
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
```

```python
# Create List for target and feature columns
target_variable = 'price'
feature_columns = []
for column in standardized_df.columns:
    if column != 'price':
        feature_columns.append(column)
        
# Create dataset for training
X = standardized_df[feature_columns]
y = standardized_df[target_variable]

# Split training data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, shuffle=True, test_size = 0.2)

# Confirm dataset splitting
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Define RandomForestRegressor training algorithm
regressor = RandomForestRegressor(verbose=1)

# Create training pipeline
pipeline = Pipeline([('scaler', MinMaxScaler()), ('regressor', regressor)])

# Train using pipeline
pipeline.fit(X_train, y_train)

# Test the trained model
predictions = pipeline.predict(X_test)

# Perform cross-validation and obtain accuracy scores
scores = cross_val_score(pipeline, X_train, y_train, cv=5)

# Display the accuracy scores
print("Accuracy scores:", scores)
print("Mean accuracy:", scores.mean())
```

```python
mean = round((scores.mean() * 100.00), 2)
print("The overall accuracy of the pipeline/model is " + str(mean) + "%.")
```
